```python
import numpy as np
import pandas as pd
```

```python
wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data')
```

```python
wine.columns = ['class','Alcohol','Malic acid','Ash', 'Alcalinity of ash','Magnesium','Total phenols','Flavanoids',
                'Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
```

```python
wine.tail()
```

```python
wine.iloc[:3,1]=np.nan
wine.head()
```

```python
wine.iloc[2:4,5]=np.nan
wine.head()
```

```python
wine['Alcohol'] = wine['Alcohol'].fillna(10)
wine['Magnesium'] = wine['Magnesium'].fillna(100)
wine.head()
```

```python
wine.isnull().sum()
```

```python
wine.dropna(inplace = True)
wine.head()
```
